### 1. requests 라이브러리 사용법

```python
# 라이브러리 불러오기
import requests as req
# 서버에게 데이터를 요청 및 응답받기 위한 라이브러리
```

```python
res = req.get('https://www.naver.com/')
# Response [200]번대 : 성공적으로 요청 및 응답이 이루어졌다는 의미
# Respense [400]번대 : 클라이언트(요청) 부분에서 오류가 발생했다는 의미
# Response [500]번대 : 서버(응답) 부분에서 오류가 발생했다는 의미
```

```python
res.text
```

### 2. BeautifulSoup 라이브러리 사용법

```python
# 라이브러리 불러오기
from bs4 import BeautifulSoup as bs
# 컴퓨터가 이해할 수 있는 파이썬 객체 형태로 변환해주는 라이브러리
```

```python
# bs(parsing할 html 문서, 'parsing 방법') ---> 객체화
# lxml : 빠르고 관대함
soup = bs(res.text,'lxml')
soup
```

```python
res = req.get('https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=')
```

```python
res
```

```python
soup = bs(res.text,'lxml')
soup
```

```python
# 태그의 위치를 찾는 방법
# 1. 내가 가져오고 싶은 contents 위에 마우스 우클릭 -> 검사
# 2. 개발자모드 좌측 상단 커서 모양 아이콘 클릭 후 가져오고 싶은 컨텐츠 클릭
```

##### 예제) 네이버 페이지에서 정보를 가져와보기
- 뉴스, 인플루언서, 통합 등과 같은 글자를 수집해보자

```python
# a태그를 가져오는 방법
soup.select('a')
# 페이지의 모든 a 태그를 가져왔음
```

```python
# a태그 중에서 클래스가 tab인 애들만 데려와보자
naver_tab = soup.select('a.tab')
naver_tab
```

```python
# 가져온 것 중에 '통합'이라는 contents를 가져와보자
```

```python
# select로 가져온 list 인덱싱 해서 가져오기
naver_tab[0]
```

```python
# select_one : 여러가지 요소 중에서 첫 번째 요소만 단수형으로 가져온다.
# select : 여러가지 요소를 복수형(list)으로 가져온다.
soup.select_one('a.tab')
```

```python
naver_tab[0].text
```

### 데이터 수집 실습
- 네이버에 '날씨'라고 검색한 후 온도데이터 수집

```python
# 현재 온도 33.3º
# 1. 페이지 정보 받아오기
# 2. 객체화 진행( 컴퓨터가 이해하기 쉬운 파이썬 객체로 변환)
# 3. 태그 선택하기 (어떤 태그를 선택할지 잘 선택하는 것이 중요!)
# 4. 컨텐츠 추출하기
```

```python
# 1. 페이지 정보 받아오기
res = req.get('https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=%EB%82%A0%EC%94%A8&oquery=&tqi=iLtfbsprvTVssgm6DY8ssssssER-124320')
```

```python
# 2. 객체화 진행( 컴퓨터가 이해하기 쉬운 파이썬 객체로 변환)
soup = bs(res.text,'lxml')
soup
```

```python
# 3. 태그 선택하기 (어떤 태그를 선택할지 잘 선택하는 것이 중요!)
soup.select('strong')[10]
# 애로 찾으면 뽑고 인덱스 세면서 찾아서 인덱싱 해야함
```

```python
# 태그 선택을 잘 하자!!!
# (기왕이면 클래스 선택자가 있는 태그를 쓰자(복붙))
# soup.select_one('div.temperature_text(띄어쓰기)span') - 계층 선택자 가져오기도 가능
weather = soup.select_one('div.temperature_text')
weather
```

```python
# 4. 컨텐츠 추출하기 .text
weather.text 
```
