## 멜론 차트 TOP100 수집하기

#### 1. 라이브러리 불러오기

```python
import requests as req # 서버와의 요청 및 응답을 위한 라이브러리
from bs4 import BeautifulSoup as bs # 컴퓨터가 이해하기 쉬운 객체로 변환해주는 라이브러리
```

#### 2. 페이지 정보 받아오기

```python
url ='https://www.melon.com/chart/index.htm'
res = req.get(url)
res
# 406(요청 실패)
# -> 멜론페이지는 보안 문제 떄문에 파이썬 코드로 접근하는 걸 막고 있다!! -> 우회접속 해야함
```

```python
# 1. 400번대 에러 발생
# 2. 접근하려는 페이지에서 F12(개발자모드)
# 3. NETWORK 탭 이동
# 4. F5 눌러서 새로고침
# 5. 맨 위에 있는 document 타입의 html 문서 정보 열람
# 6. headers 탭의 맨 아래에  있는 User-Agent 값을 복사
# 7. 딕셔너리 형태로 파이썬 코드에 붙여넣기
# 8. 붙여넣은 결과를 url 옆에 headers 값으로 지정

# h ={'User-Agent' : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"}
# req.get(url , headers = h)
```

```python
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
hdr ={'User-Agent': user_agent}
```

```python
res = req.get(url , headers = hdr )
res
```

#### 3. 객체화 진행( 컴퓨터가 이해하기 쉬운 파이썬 객체로 변환)

```python
# bs 객체화 = bs(parsing할 html문서, 'parsing할 방법')
soup = bs(res.text,'lxml')
soup
```

```python
# 노래 가수
aSinger = soup.select('div.rank02')
a =[]
for i in range(len(aSinger)):
    b = (str)(aSinger[i].text.strip('\n'))
    c = b[:len(b)//2]
    a.append(c)
for i in range(0,100):
    print(f'{i+1}. {a[i]}')
```

#### 5. DataFrame화

```python
# DataFrame 형식으로 만들어보자!
# 순위까지 추가해서!
ranklist= [] # 순위리스트
for i in range(len(song)):
     ranklist.append(i+1)
ranklist
```

```python
import pandas as pd # 데이터프레임 형식으로 만들기 위한 라이브러리
```

```python
dic= {'노래 제목': songlist,
'아티스트': singerlist,
'순위': ranklist}
dic
```
