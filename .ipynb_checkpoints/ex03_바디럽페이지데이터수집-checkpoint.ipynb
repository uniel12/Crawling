##### 데이터 전처리
- 앞 뒤 공백 제거
- 순수한 리뷰 데이터만 출력(네이버 페이 구매평이라는 글자 제거)

```python
# 앞, 뒤 공백을 업애주는 메소드 -> strip()
review[0].text.strip()
```

```python
temp = review[0].text.strip()
# 'X' in string -> string 안에 'X'라는 글자가 있다면 True, 없다면 False
'네이버 페이 구매평' in temp
```

```python
# 슬라이싱을 통해 접근
# 리뷰 내용이 매번 달라지므로 앞에서부터 접근X , 뒤에서부터 접근!
temp[-38:]
```

```python
if '네이버 페이 구매평' in temp:
    temp = temp.replace(temp[-38:],'')
print(temp)
```

```python
# 반복문 실습

reviewlist = []
for  i in range(len(review)):
    temp = review[i].text.strip()
    if '네이버 페이 구매평' in temp:
        temp = temp.replace(temp[-38:],'')
    reviewlist.append(temp)

for i in range(len(reviewlist)) :
    print(f"{i+1}. {reviewlist[i]}\n")
```

```python
# 윤석's 뒤집어서 찾기 방법
r_list=[]
for i in range(len(review)) :
    ele = review[i].text.strip() # 글자의 공백을 제거한 문자열
    if ele[::-1][0:2] == ')평' : # 문자열을 뒤집에서 시작이')평'인지 확인
        k = ele[::-1].index('(') # 시작이 ')평' 이라면 네이버 구매평이 있다고 판단 '('의 인덱스 위치 확인
        ele = ele[::-1][k+1:][::-1] # '(' 이후 글자부터 모두 슬라이싱 후 다시 뒤집기
    r_list.append(ele)
for i in range(len(r_list)) :
    print(f"{i+1}. {r_list[i]}\n")
```

```python
# 반복문 실습 (선생님 버전)
```

```python
# 1페이지만 수집하는 코드
url ='https://review4.cre.ma/bodyluv.kr/products/reviews?product_code=473&iframe_id=crema-product-reviews-2&widget_style=&app=0&parent_url=https%3A%2F%2Fbodyluv.kr%2Fproduct%2Fdetail.html%3Fproduct_no%3D473&nonmember_token=&secure_device_token=V2171dc66180b834716cabe89392e787eb80a7f2aa1084476b5b1dbaae31acb6298e3b82b31fe38664311a7eb2ea44c469&iframe=1'
res = req.get(url)
res
soup = bs(res.text,'lxml')

review = soup.select('div.review_list_v2__message.js-translate-text')
review_list =[] # 전처리된 데이터를 담을 리스트 생성
for i in range(0,10,2): # 1,3,5,7,9 등 홀수 리뷰만 출력되도록 반복
    naver_re = review[i].text.strip() 
    # 앞,뒤 공백을 없앤 리뷰 데이터를 string 형태로 저장
    if '네이버 페이 구매평' in naver_re: # 네이버페이 구매평이라는 글자가 있다면!
        naver_re = naver_re.replace(naver_re[-38:],'') # 필요없는 글자 삭제
    review_list.append(naver_re)
review_list
```

```python
# 반복문을 활용해서 1페이지부터 10페이지까지 데이터 수집
review_list =[] # 전처리된 데이터를 담을 리스트 생성
for page_num  in tq(range(1,11)):     
    url =f'https://review4.cre.ma/bodyluv.kr/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-2&page={page_num}&parent_url=https%3A%2F%2Fbodyluv.kr%2Fproduct%2Fdetail.html%3Fproduct_no%3D473&product_code=473&secure_device_token=V21bfede01de676fb1a9359c037dcec734e5a0305e28b412f6763d0bb677bed4b489db4dd30e2019fdca70033f056e81b4&widget_env=100&widget_style='
    res = req.get(url)
    res
    soup = bs(res.text,'lxml')

    review = soup.select('div.review_list_v2__message.js-translate-text')
    for i in range(0,10,2): # 1,3,5,7,9 등 홀수 리뷰만 출력되도록 반복
        naver_re = review[i].text.strip() 
        # 앞,뒤 공백을 없앤 리뷰 데이터를 string 형태로 저장
        if '네이버 페이 구매평' in naver_re: # 네이버페이 구매평이라는 글자가 있다면!
            naver_re = naver_re.replace(naver_re[-38:],'') # 필요없는 글자 삭제
        review_list.append(naver_re)
review_list
```

```python
# (참고)크롤링 오래걸리니까 반복문이 진행되는 과정을 바로 보여주는 것
# -> tq(반복문 돌리는 범위)
from tqdm import tqdm_notebook as tq
```
